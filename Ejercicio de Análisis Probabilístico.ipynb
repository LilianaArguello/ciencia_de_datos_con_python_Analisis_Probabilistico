### Responder pregunta
¿Cuánto es la duración promedio de los viajes en bicicleta?

```python
# Como tenemos una combinación de texto y número, primero hay que separarlos
ride_sharing['duration'] = ride_sharing['duration'].str.strip('minutes')
```

```python
ride_sharing
```

```python
# Dentor del mismo método de strip podemos calcular directamente el promedio
duracion_promedio = ride_sharing['duration'].str.strip('minutes').astype(int).mean()
```

```python
print(str(round(duracion_promedio,2)) + ' minutos')
```

```python
ride_sharing[((ride_sharing['user_birth_year']<1912) | (ride_sharing['user_birth_year']>2018))]
```

```python
edades_limpias = 2020 - ride_sharing[
                                    ((ride_sharing['user_birth_year']>1912) & (
                                        ride_sharing['user_birth_year']<2018))]['user_birth_year']
```

```python
edades_limpias.head(10)
```

```python
# Distribución de las edades 
plt.hist(edades_limpias)
plt.xlabel('Edad [años]')
plt.ylabel('Número de datos')
```

```python
edades_limpias.mean()
```

## Probabilidad frecuentista

Qué tan probable es que un usuario sea menor de 47 años?

```python
# Cantidad de personas debajo de 47 años
menor_47 = (edades_limpias < 47).sum()
print(menor_47)
total_personas = len(edades_limpias)
print(total_personas)
# Fraccion de personas menores a 47
probabilidad = menor_47/total_personas
print(probabilidad)
```
